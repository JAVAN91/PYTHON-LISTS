```python
with open('Uganda_Karamoja_Subcounty_Crop_Yield_Population.csv','r') as f:
    Museveni_ug = pd.read_csv("Uganda_Karamoja_Subcounty_Crop_Yield_Population.csv")
%sql PERSIST Museveni_ug;
```

```python
#Display a list of Sub Counties and their population and areas
%%sql
SELECT SUBCOUNTY_NAME, "POP", "Area"
FROM museveni_ug
```

```python
#Sort the list of districts by total crop area (descending order)
%%sql
SELECT DISTRICT_NAME
FROM museveni_ug
ORDER BY Crop_Area_Ha desc
```

```python
#Select only the Sub counties from the Moroto dist
#rict, order them alphabetically and show their production of sorghum.
%%sql
SELECT SUBCOUNTY_NAME FROM museveni_ug
WHERE DISTRICT_NAME= 'MOROTO'
ORDER BY "SUBCOUNTY_NAME" ASC
```

```python
#Compute the total Maize production per District.
%%sql
SELECT District_Name,count(M_Yield_Ha)
FROM museveni_ug
GROUP BY DISTRICT_NAME
```

```python
#Compute the number of Sub counties where Maize is
# produced and the total Maize production per District.
%%sql
SELECT ("DISTRICT_NAME"),Count(M_Prod_Tot) FROM museveni_ug
ORDER BY DISTRICT_NAME
```

```python
#Compute the overall Crop area in all Sub counties 
#where population is over 20000.
%%sql
SELECT POP,SUBCOUNTY_NAME,Crop_Area_Ha 
FROM museveni_ug
WHERE"POP">20000
```

```python
#Sort the Maize production in descending order by Districts, only taking into account Sub counties where Maize area is
 # than Sorghum area, and display the number of Sub counties per district matching that criteria.
 %%sql
 SELECT DISTRICT_NAME ,M_AREA_Ha,S_Area_Ha 
 FROM museveni_ug
 WHERE "M_AREA_Ha">"S_Area_Ha"
 ORDER BY "M_Prod_Tot" DESC;
```
